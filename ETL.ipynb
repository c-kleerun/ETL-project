```python
import pandas as pd
from sqlalchemy import create_engine
from config import user
from config import password
```

```python
tomato = "Resources/rotten_tomatoes.csv"
imdb = "Resources/IMDB_movies.csv"
```

```python
tomato_df = pd.read_csv(tomato)
imdb_df = pd.read_csv(imdb)
```

```python
tomato_df.head()
```

```python
imdb_df.head()
```

```python
imdb_clean_df = imdb_df[['original_title', 'avg_vote', 'usa_gross_income']]
imdb_clean_df.head()
```

```python
tomato_clean_df = tomato_df.drop(columns='No. of Reviews')
tomato_clean_df.head()
```

```python
tomato_clean_df['Title'] = tomato_clean_df['Title'].str.split(r'(', expand=True)[0].str.strip().str.lower()
tomato_clean_df
```

```python
imdb_clean_df = imdb_clean_df.rename(columns={"original_title": "title", "avg_vote": "imdb_score", "usa_gross_income": "usa_gross"})
imdb_clean_df.head()
```

```python
tomato_clean_df = tomato_clean_df.rename(columns={"Rank": "rank", "Title": "title", "RatingTomatometer": "rt_rating", "Genres": "genres"})
tomato_clean_df.head()
```

```python
clean_score = imdb_clean_df["imdb_score"]*10
imdb_clean_df['imdb_score'] = clean_score
imdb_clean_df.head()
```

```python
imdb_clean_df['title'] = imdb_clean_df['title'].str.strip().str.lower()
imdb_clean_df
```

```python
imdb_clean_df.to_csv('imdb_clean.csv', index=False)
```

```python
tomato_clean_df.to_csv('tomato_clean.csv', index=False)
```

```python
engine = create_engine(f'postgresql://{user}:{password}@localhost:5432/ETL_project')
conn = engine.connect()
```

```python
imdb_clean_df.to_sql('imdb', con=engine, index=False, if_exists='replace')
```

```python
tomato_clean_df.to_sql('tomato', con=engine, index=False, if_exists='replace')
```

```python
combined = pd.read_sql("""SELECT t.title, t.rt_rating, i.imdb_score, i.usa_gross
    FROM tomato AS t 
    JOIN imdb AS i ON 
    i.title = t.title""", conn)
combined
```

```python
combined_cleaned = combined.drop_duplicates(subset=['title']) 
combined_cleaned
```

```python
combined_sorted = combined_cleaned.sort_values(['rt_rating', 'imdb_score'], ascending=False)
combined_sorted.head(10)
```

```python
combined_sorted_imdb = combined_cleaned.sort_values(['imdb_score', 'rt_rating'], ascending=False)
combined_sorted_imdb.head(10)
```

```python
combined_sorted_imdb = combined_cleaned.sort_values(['imdb_score', 'rt_rating', 'usa_gross'], ascending=False)
combined_sorted_imdb.head(10)
```
